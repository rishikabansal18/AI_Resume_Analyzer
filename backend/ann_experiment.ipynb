```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = []
y = []

for _ in range(500):
    similarity = np.random.rand()
    skill_match = np.random.rand()

    # scoring logic (same as backend)
    score = 0.6 * similarity + 0.4 * skill_match

    X.append([similarity, skill_match])
    y.append(score)

X = np.array(X)
y = np.array(y)

print("Sample X:", X[:5])
print("Sample y:", y[:5])
```

```python
model = Sequential([
    Dense(8, activation='relu', input_shape=(2,)),
    Dense(4, activation='relu'),
    Dense(1, activation='sigmoid')   # 🔥 important
])
```

```python
model.compile(
    optimizer='adam',
    loss='mse'
)
```

```python
history = model.fit(X, y, epochs=50, verbose=1)
```

```python
test_input = np.array([[0.7, 0.5]])
prediction = model.predict(test_input)

print("Predicted Score:", prediction[0][0])
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.title("Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```
